```python
import pandas as pd

print(pd.__version__)
```

Series: A 1-Dimensional labeled array that can hold any datatype. It can be taken like a single column in a spreadsheet.

```python
data = [1011 , 1022, 100]


series = pd.Series(data)          # Series() is a constructor and not a function

print(series)
```

```python
data = [1011 , 1022, 100]


series = pd.Series(data, index = {"a" , "b" , "c"})          # we can pass list , tuple, dictionary , set , a numpy array or even another series as index
print(series)

series.loc["a"] =200

print(series.loc["a"])  

print(series.iloc[0])
```

```python
data = [1011 , 1022, 100, 200 , 750 , 800, 670, 69]

series = pd.Series(data, index = ['a' , 'b' , 'c' , 'd' , 'e', 'f' , 'g' , 'h'])

print(series[series >= 200])
```

```python
calories = {"Day 1" : 1070,
            "Day 2" : 980,
            "Day 3" : 2350,
            "Day 4" : 500}

 
series = pd.Series(calories)  # No need to pass index for dictionary because key acts as an index.

print(series)
print("-----------------------------------------------------")
print(series.loc["Day 2"])
print("-----------------------------------------------------")
print(series.iloc[3])
print("-----------------------------------------------------")
series.loc["Day 4"] += 567

print(series.loc["Day 4"])
print("-----------------------------------------------------")
print(series)
```

```python
calories = {"Day 1" : 1070,
            "Day 2" : 980,
            "Day 3" : 2350,
            "Day 4" : 500}

 
series = pd.Series(calories)

print(series[series>2000])
```

Dataframe : A dataframe is a tabular data structure with rows and columns. It is 2 dimensional like a excel spreadsheet

```python
data = {"Name" : ["Pikachu" , "Meowth" , "Charizard" , "Dragonite" , "Mew" , "Goodra"],
        "Post" : ["a", "b" , "c", "d", "e", "f"],
        "age": [11,2,3,4]}


print(data)

print(data["Name"])
```

```python
data = {"Name" : ["Pikachu" , "Meowth" , "Charizard" , "Dragonite" , "Mew" , "Goodra"],
        "Pokedex No.": [25,52,6,149,151,706]}

df = pd.DataFrame(data)

print(df)
print('----------------------------------------')
print(df.head())
print('----------------------------------------')
print(df.tail())
print('----------------------------------------')

```

```python
data = {"Name" : ["Pikachu" , "Meowth" , "Charizard" , "Dragonite" , "Mew" , "Goodra"],
        "Pokedex No.": [25,52,6,149,151,706]}

df = pd.DataFrame(data , index = ['Pokemon 1','Pokemon 2','Pokemon 3','Pokemon 4','Pokemon 5','Pokemon 6'])

print(df)
print('----------------------------------------')
print(df.loc["Pokemon 4"])
print('----------------------------------------')
print(df.iloc[2])
print('----------------------------------------')
print(df.iloc[1,1])      # Access value at index number 1 and column number 1
print('----------------------------------------')
print(df.iat[2,0])            # Same as iloc just faster
print('----------------------------------------')



value1 = df.loc[df["Name"] == 'Pikachu', 'Pokedex No.']
print(value1)
print('----------------------------------------')
value2 = df.loc[df["Name"] == 'Pikachu', 'Pokedex No.'].item()

value3 = df.loc[df["Name"] == 'Pikachu', 'Pokedex No.'].values[0]

print(value2, value3)

```

```python
data = {"Name" : ["Pikachu" , "Meowth" , "Charizard" , "Dragonite" , "Mew" , "Goodra"],
        "Pokedex No.": [25,52,6,149,151,706]}

df = pd.DataFrame(data , index = ['Pokemon 1','Pokemon 2','Pokemon 3','Pokemon 4','Pokemon 5','Pokemon 6'])

print(df)
print('----------------------------------------')

value = df.loc[df["Pokedex No."] == 52 ]
print(value)

print('----------------------------------------')

value = df.loc[df["Pokedex No."] == 52 , 'Name'].item()
print(value)
```

```python
# Add a new column

data = {"Name" : ["Pikachu" , "Meowth" , "Charizard" , "Dragonite" , "Mew" , "Goodra"],
        "Pokedex No.": [25,52,6,149,151,706]}

df = pd.DataFrame(data , index = ['Pokemon 1','Pokemon 2','Pokemon 3','Pokemon 4','Pokemon 5','Pokemon 6'])

print(df)

print('---------------------------------------------------------')


df['Type'] = ['Electric','Normal','Fire/Flying','Dragon/Flying','Psychic','Dragon']

print(df)

print('---------------------------------------------------------')

# Add a new row : Easiest way to add row is to create a dataframe and concatenate.

new_row = pd.DataFrame([{"Name": "Sylveon" , "Pokedex No.":700 , "Type" : "Fairy" }] , index = ["Pokemon 7"])

df = pd.concat([df, new_row])

print(df)

```

```python
# Add new rows
new_rows = pd.DataFrame([{"Name": "Darkrai" , "Pokedex No.":491 , "Type" : "Dark" },
                         {"Name": "Gengar" , "Pokedex No.":94 , "Type" : "Ghost/Poison"},
                         {"Name": "Pheromosa" , "Pokedex No.":795 , "Type" : "Bug/Fighting"}] , index = ["Pokemon 8" , "Pokemon 9" , "Pokemon 10"])

df = pd.concat([df, new_rows])

print(df)
```

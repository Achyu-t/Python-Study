Series: A 1-Dimensional labeled array that can hold any datatype. It can be taken like a single column in a spreadsheet.

```python
data = [1011 , 1022, 100]


series = pd.Series(data)          # Series() is a constructor and not a function

print(series)
```

```python
data = [1011 , 1022, 100]


series = pd.Series(data, index = {"a" , "b" , "c"})          # we can pass list , tuple, dictionary , set , a numpy array or even another series as index
print(series)

series.loc["a"] =200

print(series.loc["a"])  

print(series.iloc[0])
```

```python
data = [1011 , 1022, 100, 200 , 750 , 800, 670, 69]

series = pd.Series(data, index = ['a' , 'b' , 'c' , 'd' , 'e', 'f' , 'g' , 'h'])

print(series[series >= 200])
```

```python
calories = {"Day 1" : 1070,
            "Day 2" : 980,
            "Day 3" : 2350,
            "Day 4" : 500}

 
series = pd.Series(calories)  # No need to pass index for dictionary because key acts as an index.

print(series)
print("-----------------------------------------------------")
print(series.loc["Day 2"])
print("-----------------------------------------------------")
print(series.iloc[3])
print("-----------------------------------------------------")
series.loc["Day 4"] += 567

print(series.loc["Day 4"])
print("-----------------------------------------------------")
print(series)
```

```python
calories = {"Day 1" : 1070,
            "Day 2" : 980,
            "Day 3" : 2350,
            "Day 4" : 500}

 
series = pd.Series(calories)

print(series[series>2000])
```

```python
data = {"Name" : ["Pikachu" , "Meowth" , "Charizard" , "Dragonite" , "Mew" , "Goodra"],
        "Pokedex No.": [25,52,6,149,151,706]}

df = pd.DataFrame(data , index = ['Pokemon 1','Pokemon 2','Pokemon 3','Pokemon 4','Pokemon 5','Pokemon 6'])

print(df)
print('----------------------------------------')
print(df.loc["Pokemon 4"])
print('----------------------------------------')
print(df.iloc[2])
print('----------------------------------------')
print(df.iloc[1,1])      # Access value at index number 1 and column number 1
print('----------------------------------------')
print(df.iat[2,0])            # Same as iloc just faster
print('----------------------------------------')



value1 = df.loc[df["Name"] == 'Pikachu', 'Pokedex No.']
print(value1)
print('----------------------------------------')
value2 = df.loc[df["Name"] == 'Pikachu', 'Pokedex No.'].item()

value3 = df.loc[df["Name"] == 'Pikachu', 'Pokedex No.'].values[0]

print(value2, value3)

```

```python
data = {"Name" : ["Pikachu" , "Meowth" , "Charizard" , "Dragonite" , "Mew" , "Goodra"],
        "Pokedex No.": [25,52,6,149,151,706]}

df = pd.DataFrame(data , index = ['Pokemon 1','Pokemon 2','Pokemon 3','Pokemon 4','Pokemon 5','Pokemon 6'])

print(df)
print('----------------------------------------')

value = df.loc[df["Pokedex No."] == 52 ]
print(value)

print('----------------------------------------')

value = df.loc[df["Pokedex No."] == 52 , 'Name'].item()
print(value)
```

Aggregate functions: Functions that reduces set of values  into a single summary value. Used to summarize and analyze data.

Often used with groupby() function.

```python
# Aggregate Functions for Whole datasets.

df = pd.read_csv(r"D:\Python training\datasets\data_pkmn.csv")

# print(df.mean(numeric_only = True))   # Find the mean of any columns that are numeric

df.sum(numeric_only = True)

df.min(numeric_only = True)

df.max(numeric_only = True)

df.count()         # NaN is ignored by count. So they won't count
```

```python
# Column Specific aggregate functions

print(df["Height"].mean())   

print(df["Weight"].sum())

print(df['Height'].min())

print(df['Weight'].max())

print(df["Type2"].count())
```

```python
# Group by

group = df.groupby("Type1")

# print(group["Height"].mean())

group["Height"].sum()

group["Height"].min()

group["Weight"].max()

group['Name'].count()
```

Handle Missing and Inconsistent Data : 

```python
df = pd.read_csv(r"D:\Python training\datasets\data_pkmn.csv")

# Drop missing values
df_new= df.dropna(subset = ["Type2"])
df_new.head(10)

# Replace missing values
df_new2 = df.fillna({"Type2" : "None"})
df_new2.head(15)

# Fix inconsistent values

df["Type1"] = df["Type1"].replace({"Grass" : "GHAS", "Fire" : "AAGH", "Water" : "PANI"})
df.head(10)


# Standardize text

df["Name"] = df["Name"].str.lower()
df.head(10)

# Fix or chnage Data Types

df["Legendary"] = df["Legendary"].astype(bool)
df.tail(10)

# Remove duplicate Values
df.head()

df = df.drop_duplicates()

df.head()
```

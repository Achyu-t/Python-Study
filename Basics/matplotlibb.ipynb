```python
import matplotlib.pyplot as plt
```

Plot Customization

```python
x = [2023, 2024 , 2025 , 2026]

y1= [44,34,56,90]
y2 = [46,766,21,13]
y3 = [19,67,69,10]

line_style = dict(marker= '.'  , ms =20 , markerfacecolor = "Black" , mec = "Pink" , linestyle = 'dashed' , linewidth = 4 )

plt.plot(x, y1 , color = 'pink', **line_style)    

plt.plot(x, y2 , color = 'green' , **line_style)  

plt.plot(x , y3, color = 'blue', **line_style)

plt.show()                                                                                                                                                                                                         
```

# Some common markers are : v , . , , , 1 , 2 , 3 , 4 , 8 , s , P , p , x , h , H , < , > , + , * , ^

# ms = markersize . We can use any of the two as keyword.
# mfc also works. It takes colours in hex or name.
# mec = Marker edge colour.
# linestyle = default is a solid line,  dotted , dashdot , None (opnly markers).


Lables and Titles:

```python
import numpy as np

x = np.array([2023,2024,2025,2026])

y1= np.array([10,20,30,40])
y2 = np.array([15,25,35,45])
y3= np.array([12,22,32,43])

plt.title("Amount Spent each Year in Million" , fontsize = 20 , 
                                                family = "Times New Roman" , 
                                                fontweight = 'bold', 
                                                color = 'pink')

plt.xlabel("Year" , fontsize = 20)
plt.ylabel("Money Spent" )
plt.plot(x , y1)
plt.plot(x , y2)
plt.plot(x , y3)

# plt.show()

```

```python
import matplotlib.pyplot as plt
```

```python
x = [2023, 2024 , 2025 , 2026]

y= [44,34,56,90]

plt.plot(x, y , marker= '.'  , ms =20 , markerfacecolor = "Black" , mec = "Pink" , linestyle = 'dashed' , linewidth = 4 , color = "Orange")    
plt.show()                                                                                                                                                                                                         
```

# Some common markers are : v , . , , , 1 , 2 , 3 , 4 , 8 , s , P , p , x , h , H , < , > , + , * , ^

# ms = markersize . We can use any of the two as keyword.
# mfc also works. It takes colours in hex or name.
# mec = Marker edge colour.
# linestyle = default is a solid line,  dotted , dashdot , None (opnly markers).

Plot Customization

```python
x = [2023, 2024 , 2025 , 2026]

y1= [44,34,56,90]
y2 = [46,766,21,13]
y3 = [19,67,69,10]

line_style = dict(marker= '.'  , ms =20 , markerfacecolor = "Black" , mec = "Pink" , linestyle = 'dashed' , linewidth = 4 )

plt.plot(x, y1 , color = 'pink', **line_style)    

plt.plot(x, y2 , color = 'green' , **line_style)  

plt.plot(x , y3, color = 'blue', **line_style)

plt.show()                                                                                                                                                                                                         
```

# Some common markers are : v , . , , , 1 , 2 , 3 , 4 , 8 , s , P , p , x , h , H , < , > , + , * , ^

# ms = markersize . We can use any of the two as keyword.
# mfc also works. It takes colours in hex or name.
# mec = Marker edge colour.
# linestyle = default is a solid line,  dotted , dashdot , None (opnly markers).


Lables and Titles:

```python
import numpy as np

x = np.array([2023,2024,2025,2026])

y1= np.array([10,20,30,40])
y2 = np.array([15,25,35,45])
y3= np.array([12,22,32,43])

plt.title("Amount Spent each Year in Million" , fontsize = 20 , 
                                                family = "Times New Roman" , 
                                                fontweight = 'bold', 
                                                color = 'pink')

plt.xlabel("Year" , fontsize = 20 , family = 'Arial', fontweight = 'bold')
plt.ylabel("Money Spent" )

plt.tick_params(axis ='both' , colors = 'green')      # axis can be x , y or both


plt.plot(x , y1)
plt.plot(x , y2)
plt.plot(x , y3)

plt.xticks(x)


plt.show()

```

```python
# Grid Lines :

x = [1,2,3,4,5]
y= [5,10,15,20,25]

# plt.grid()

plt.grid(axis = 'x' , linewidth = 3 , color = "cyan" , linestyle = "dashed")

plt.plot(x,y)
plt.show()
```

Bar Charts

```python
import numpy as np

categories = ["Grains" , "Fruits" , "Vegetables" , "Protein" , "Dairy" , "Sweets"]

values=np.array([4,3,2,5,3,1])

# plt.bar(categories , values , color = "pink")

plt.barh(categories , values , color = "pink")

plt.title("Daily consumption")
plt.xlabel("Food")
plt.ylabel("Amount")
plt.show()
```

Pie Chart

```python
categories = ["1st year" , "2nd year" , "3rd year" , "4th year"]
values= np.array([37,42,48,8])
colours = ["Pink" , "Red" , "Cyan" , "Magenta"]
plt.pie(values, labels = categories ,  autopct = "%1.1f%%" , colors= colours , explode = [0,0,0,0.15], shadow = True , startangle = 180)  

# Explode can be any value but if larger than 1 the piece moves way out of the circle.

plt.title("My College Student Distribution")
plt.show()
```

Scatter Graph

```python
import numpy as np

x1 = np.array([0,1,5,7,11,14,7,8,5,9]) 

y1 = np.array([55,60,65,62,77,84,67,69,63,59])

x2 = np.array([0,1,2,3,4,5,6,7,6,7]) 

y2 = np.array([56,61,45,67,89,98,78,67,91,76])

plt.scatter(x1,y1 , color = "Orange" , alpha = 0.7 , s= 100 , label = "Class A")

# alpha is used to change transparency of the dots. s means size

plt.scatter(x2,y2 , color = "Blue" , alpha = 1 , s= 100 , label = "Class B")

plt.title("Study time vs Grades Relation")
plt.xlabel("Hours studied")
plt.ylabel("Marks obtained")

plt.legend()

plt.show()
```

Histogram : A visual representation of the distribution of quantitative data. They group values into intervals, also known as bins and count how many fall in each range

```python
import numpy as np

scores = np.random.normal(loc = 80 , scale = 10 , size =100)

#  .normal() is used for Normal Distribution. 
# loc means location of median (center). loc = 80 means all test scores will cluster around 80. (most frequent score)
# scale gives standard deviation
# size  = how many numbers to generate

scores = np.clip(scores, 0, 100)   # Setting minimum and maximum number boundry. So in this case number below 0 are made 0 and above 100 are made 100.

plt.hist(scores, bins = 10, color = "purple", edgecolor = 'black')   
# bins gives number of intervals

plt.title("Exam scores")
plt.xlabel("Score")
plt.ylabel("Number of students")

plt.show()
```

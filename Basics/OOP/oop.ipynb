Inheritance: Allows a class to inherit attributes and methods from another class. Helps with code reusability and extensibility. 

```python
class Animal:
    def __init__(self, name):
        self.name = name
        self.isalive = True

    def eat(self):
        print(f"{self.name} is eating.")

    def sleep(self):
        print(f"{self.name} is sleeping.")

class Dog(Animal):
    def speak(self):
        print('WOOF')

class Cat(Animal):
    def speak(self):
        print('MEOW')

class Mouse(Animal):
    def speak(self):
        print('SQUEAK')


dog = Dog('Scooby')
cat = Cat('Sparky')
mouse = Mouse('Jerry')


print(dog.name)
print(dog.isalive)

print(cat.name)
cat.eat()

print(cat.name)
mouse.sleep()

dog.speak()
cat.speak()
mouse.speak()
```

MULTIPLE INHERITANCE : Inherit from more than one parent class =>  C(A,B)
MULTI LEVEL INHERITANCE : Inherit from a parent which inherits from another parent :  C(B) <-  B(A) <- A

```python
class Animal:

    def __init__(self,name):
        self.name = name

    def eat(self):
        print('This animal is eating')
    
    def sleep(self):
        print("This animal is sleeping")



class Predator(Animal):
    def hunt(self):
        print(f" {self.name} is hunting")


class Prey(Animal):
    def flee(self):
        print(f"{self.name} is fleeing")



class Rabbit(Prey):
    pass



class Hawk(Predator):
    pass



class Fish(Prey, Predator):
    pass



rabbit = Rabbit("Tom")
hawk = Hawk('Bazuka')
fish = Fish('Nemo')


rabbit.flee()

hawk.hunt()

fish.hunt()
fish.flee()

fish.sleep()

rabbit.eat()

print(rabbit.name)
print(fish.name)
```

SUPER FUNCTION : Function used in a child class to call methods from a parent class (superclass). It allows us to extend the functionality of inherited methods.

```python
import math
class Shape:
    def __init__(self, colour, filled):
        self.colour = colour
        self.filled = filled
    
    def describe(self):
        print(f"It is {self.colour} and {'filled' if self.filled else 'not filled'} ")


class Circle(Shape):
    def __init__(self,colour, filled, radius):
        super().__init__(colour,filled)
        self.radius = radius

    def describe(self):
        print(f"It is a circle with area : {math.pi* self.radius**2} m.sq")

class Square(Shape):
    def __init__(self,length , colour , filled):
        super().__init__(colour,filled)
        self.length = length 
    
    def describe(self):
        print(f"It is a square with area : {self.length ** 2} m.sq")

class Triangle(Shape):
    def __init__(self, height , base , colour , filled):
        super().__init__(colour,filled)
        self.height = height 
        self.base = base
        # super().describe()

    def describe(self):

        print(f"It is a triangle with area : {0.5*self.base *self.height} m.sq")
        super().describe()

circle1 = Circle("red",True,7)
square1 = Square(10,"blue",True)
triangle1 = Triangle(3,4,'green',False)


print(circle1.colour)

print(square1.length)

print(triangle1.filled)

circle1.describe()

square1.describe()

triangle1.describe()
```

Polymorphism : Greek word meaning to have many forms or faces.   Poly meaning Many and Morphe meaning Form

IN OOP there are two ways to acheive polymorphism. 

1. Inheritance 
2. Duck Typing 

```python
import math
from abc import ABC, abstractmethod

class Shape:
    @abstractmethod
    def area(self):
        pass

class Circle(Shape):
    def __init__(self,radius):
        self.radius = radius

    def area(self):
        return math.pi*self.radius**2


class Square(Shape):
    def __init__(self, length):
        self.length = length

    def area(self):
        return self.length**2


class Trianle(Shape):
    def __init__(self, height, base):
        self.height = height
        self.base = base

    def area(self):
        return 0.5*self.height*self.base
    

class Pizza(Circle):
    def __init__(self,topping,radius):
        self.topping= topping
        super().__init__(radius)   


shapes = [Circle(4), Square(5) , Trianle(3,4), Pizza("Pineapple", 22)]

for shape in shapes:
    print(f'The area is : {shape.area()} m²')

```

Duck Typing :  Based on idea that " If it looks like a duck and quacks like a duck , it must be a duck".  To acheive polymorphism, objects must have minimum necessary attributes/methods. 

```python
class Animal:
    alive = True

class Dog(Animal):
    def speak(self):
        print("Woof")

class Cat(Animal):
    def speak(self):
        print("Meow")


class Car:
    alive = False
    def speak(self):
        print('Honk')

animals = [Dog() , Cat(), Car()]

for animal in animals:
    animal.speak()
    print(animal.alive)
```

Static method : A method that belongs to a class rather than any object from that class (instance). Usually used for general utility functions that donot need access to class data

```python
class Employee:

    def __init__(self,name,job_position):
        self.name = name
        self.position = job_position

    def get_info(self):
        return f"{self.name} = {self.position}" 
    
    @staticmethod
    def is_valid_position(position):
        valid_position = ['Manager', 'Cashier' , 'Cook', 'Janitor' , 'Waiter']
        return position in valid_position


employe1 = Employee("Ram", "Manager")
employe2 = Employee("Gordon", "Cook")
employe3 = Employee("Pikachu", "pokemon")

print(Employee.is_valid_position('Cook'))

print(employe1.get_info())
print(employe3.get_info())
print(employe2.get_info())

employe1.get_info()
```

Class Methods: Methods that allow operations related to the class

class method take (cls) as first parameter just like how instance method takes (self) as a first parameter.

```python
class Student:

    count = 0

    def __init__(self, name , gpa):
        self.name = name
        self.gpa = gpa

        Student.count += 1

    # Instance Method

    def get_info(self):
        return f"{self.name}:{self.gpa}"
    

    @classmethod
    def get_count(cls):
        return f"Total number of students: {cls.count}"

student1 = Student("Harry" , 3.0)
student2 = Student("Ron" , 4.0)
student3 = Student("Hermoine" , 6.0)

print(Student.get_count())
```

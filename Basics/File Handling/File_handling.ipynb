File writing : .txt , .json , .csv

File mode : 'w' = write 
            'x' = also write but only if file doesnot exist. If file exists it returns error.
            'r' = read 
            'a' = append

```python
txt_data = " I like pokemon"
txt_data2 = " I like coding more though"

file_path = "output.txt"

output_filepath = r"C:\Users\achyu\Desktop\output.txt"


with open(file_path ,  'w') as file:  # or we can also write as :  with open(file = file_path , mode = 'w') as file: (keyword argument style if its suits you)
    file.write(txt_data)
    print(f"txt file '{file_path}' was created")

try:
    with open(output_filepath ,  'x') as file:
        file.write(txt_data)
        print(f"txt file '{file_path}' was created")

except FileExistsError:
    print("File already exists")


with open(output_filepath ,  'a') as file:
        file.write("\n" + txt_data2)
        print(f"txt file '{file_path}' was created")
```

```python
employees = ['Pikachu' , 'Charizard' , 'Mewtwo' , 'Goodra' , 'Sceptile']

file_path = r'D:\Python training\Basics\File Handling\employeedata.txt'

with open(file_path , 'w') as file:

    for employee in employees:
        file.write(employee + '\t')  # or   file.write(employee + ' ')

    print("The file was created and employee name is added")


```

```python
import json

employees = { "Name" : "Pikachu" , 
             "Role" : "Manager",
              "Age" : "10",
              "Battles Won" : "67"            
}



file_path = r'D:\Python training\Basics\File Handling\employeedata.json'

with open(file_path , 'w') as file: 
    json.dump(employees , file , indent = 4)             # Indenting (moving to right) each key value pair by four spaces for better readability.
    print(f"json file '{file_path}' was created")
```

```python
import csv

employees = [["Name" , "Age" , "Job"] , 
             ["Pikachu", 10 , "Manager"], 
             ["Goodra" , 15 , "Customer Care"], 
             ["Sceptile" , 67 , "Chef"]]

file_path = r'D:\Python training\Basics\File Handling\employeedata.csv'

with open (file_path , 'w') as file:

    writer = csv.writer(file)   # Object that provides method for writing data to a csv file.
    for row in employees:
        writer.writerow(row)
    print(f"csv file '{file_path}' was created")

```

```python
import csv

employees = [["Name" , "Age" , "Job"] , 
             ["Pikachu", 10 , "Manager"], 
             ["Goodra" , 15 , "Customer Care"], 
             ["Sceptile" , 67 , "Chef"]]

file_path = r'D:\Python training\Basics\File Handling\employeedata2.csv'

with open (file_path , 'w', newline = '') as file:  # This added paremeter prevent an extra newline space after each row.

    writer = csv.writer(file)   
    for row in employees:
        writer.writerow(row)
    print(f"csv file '{file_path}' was created")
```

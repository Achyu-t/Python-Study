Heatmap : Generally used for plotting correlation

```python
titanic_dataset = sns.load_dataset('titanic')

titanic_dataset.head()
```

```python
titanic_dataset.sex = titanic_dataset.sex.apply(lambda x : 1 if x == 'male' else 0)

titanic_dataset.alive = titanic_dataset.alive.apply(lambda x : 1 if x=='yes' else 0)

titanic_dataset.alone = titanic_dataset.alone.apply(lambda x: 1 if x else 0)
```

```python
titanic_dataset.head()
```

```python
import pandas as pd


titanic_dataset['class'] = titanic_dataset['class'].astype(pd.CategoricalDtype(categories= ['First' , 'Second' , "Third"], ordered = True)).cat.codes

titanic_dataset = titanic_dataset.drop(['embarked' , 'who' , 'adult_male' , 'deck' , 'embark_town'] , axis = 1)

titanic_dataset.head()
```

```python
# Correlation

titanic_dataset.corr()
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize = (12,8))
sns.heatmap(titanic_dataset.corr() , annot = True , cmap = 'coolwarm')
plt.tight_layout()
```
